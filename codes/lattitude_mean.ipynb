```python
from aoespy import *
import xarray as xr
dir='/nobackupp27/afahad/exp/'
data='/nobackupp27/afahad/exp/script_replay_AGU/data/'
```

```python
cntrl=xr.open_mfdataset('/nobackupp27/afahad/exp/GEOSMIT44_cntrl/holding/geosgcm_surf/200505/*surf*200505*z.nc4').sel(time=slice('2005-04','2006-04'))
rep=xr.open_mfdataset('/nobackupp27/afahad/exp/GEOSMIT41_ciauF/holding/geosgcm_surf/200505/*surf*200505*z.nc4').sel(time=slice('2005-04','2006-04'))
rep=rep.sel(time=slice('2005-05','2006-04'))
```

```python
cntrlP=(cntrl.PRECTOT.compute())*86400
repP=(rep.PRECTOT.compute())*86400
```

```python
cntrlPs=(cntrl.PS.compute())
repPs=(rep.PS.compute())
```

```python
clf=arange(0,90,5)
f()
plt.subplot(2,1,1)
cntrlP.sel(lat=slice(-2,2)).mean(dim='lat').plot(levels=clf,cmap='YlOrRd')
plt.subplot(2,1,2)
repP.sel(lat=slice(-2,2)).mean(dim='lat').plot(levels=clf,cmap='YlOrRd')
```

```python
clf=arange(-90,90,5)
f()
(cntrlP-repP).sel(lat=slice(28,32)).mean(dim='lat').plot(levels=clf,cmap='coolwarm')
```

```python
clf=arange(-125,125+1,50) #levels=clf,
i=-2
f()
(cntrlP-repP).sel(lat=slice(i,i+2)).mean(dim='lat').plot(cmap='coolwarm',extend='both')
```
