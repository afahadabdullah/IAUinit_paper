```python
# sst.sst

def shift_longitude_roll(ds, lon_name='lon'):
  roll_amount = int(180 / (ds[lon_name][1] - ds[lon_name][0]))

  ds = ds.roll({lon_name: -roll_amount}, roll_coords=True)
  # Adjust coordinates after rolling
  ds = ds.assign_coords({lon_name: xr.where(ds[lon_name] > 180, ds[lon_name] - 360, ds[lon_name])})
  return ds.sortby(lon_name)

ds_shifted = shift_longitude_roll(noaa)
noaa=ds_shifted.interp(lon=ME426.lon, lat=ME426.lat)
```
