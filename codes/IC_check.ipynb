```python
from aoespy import *
import xarray as xr
```

```python
dir='/nobackupp27/afahad/exp/'
exp1='GEOSMIT44_cntrl'
exp2='GEOSMIT41_ciauF'
```

```python
ic1=xr.open_dataset(dir+exp1+'/holding/geosgcm_surf/200505/GEOSMIT44_cntrl.geosgcm_surf.20050501_0130z.nc4')
ic2=xr.open_dataset(dir+exp2+'/holding/geosgcm_surf/200505/GEOSMIT41_ciauF.geosgcm_surf.20050428_0130z.nc4')
```

```python
f()
plt.subplot(2,2,1)
plot(squeeze(ic1.FRSEAICE));
plt.title('MERRA2-ECCO IC',fontsize=12,fontweight='bold')
plt.subplot(2,2,2)
plot(squeeze(ic2.FRSEAICE));
plt.title('IAU IC',fontsize=12,fontweight='bold')
plt.subplot(2,1,2)
plot(squeeze(ic1.FRSEAICE)-squeeze(ic2.FRSEAICE),clf=arange(-.45,.451,.1));
plt.title('MERRA2-ECCO minus IAU IC',fontsize=12,fontweight='bold')
plt.savefig('FRSEAICE_IC.png',dpi=200)
```
